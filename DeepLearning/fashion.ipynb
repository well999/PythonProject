```python
from matplotlib import pyplot as plt
from torchvision import transforms
from torchvision import datasets
%matplotlib inline
```

```python
transform = transforms.Compose(
    [
        transforms.Resize([224, 224]),
        transforms.ToTensor(),
    ]
)
mnist_train = datasets.FashionMNIST("./data", train=True, transform=transform)
mnist_test = datasets.FashionMNIST("./data", train=False, transform=transform)
```

```python
fig, axes = plt.subplots(10, 10, sharex=True, sharey=True, dpi=120, figsize=(6, 6))
for i in range(10):
    for j in range(10):
        axes[i, j].imshow(mnist_test[i * 10 + j][0][0].numpy(), cmap=plt.cm.gray_r)
        axes[i, j].set_axis_off()
fig.tight_layout()
```
